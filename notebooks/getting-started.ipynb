```python
import os
from velixar import Velixar

# Set your API key
os.environ["VELIXAR_API_KEY"] = "vlx_your_key_here"  # Replace with your key
v = Velixar()
print("Connected!")
```

## Store Memories
Store different types of information — preferences, facts, decisions, goals.

```python
memories = [
    ("User prefers dark mode and Python for backend", ["preferences"]),
    ("Working on a RAG pipeline using LangChain and Qdrant", ["project"]),
    ("Decided to use serverless architecture on AWS Lambda", ["decision"]),
    ("Learning about knowledge graphs and entity extraction", ["learning"]),
    ("Goal: launch MVP by end of March", ["goals"]),
]

stored = []
for content, tags in memories:
    result = v.store(content, tags=tags)
    stored.append(result)
    print(f"Stored: {result.id} | {content[:50]}...")
```

## Semantic Search
Velixar finds relevant memories even when the query uses different words.

```python
queries = [
    "what tech stack is the user working with?",
    "what are their goals and deadlines?",
    "what UI preferences do they have?",
]

for q in queries:
    print(f"
Query: {q}")
    results = v.search(q, limit=2)
    for m in results:
        print(f"  [{m.score:.2f}] {m.content}")
```

## List All Memories

```python
all_memories = v.list(limit=20)
print(f"Total: {len(all_memories)} memories
")
for m in all_memories:
    print(f"{m.id[:8]}... | {str(m.tags):30s} | {m.content[:60]}")
```

## Update and Delete

```python
# Update tags on first memory
if stored:
    v.update(stored[0].id, tags=["preferences", "ui", "updated"])
    print(f"Updated tags on {stored[0].id}")

    # Delete last memory
    v.delete(stored[-1].id)
    print(f"Deleted {stored[-1].id}")
```
